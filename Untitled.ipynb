```python
n = 200
mus1 = np.array([0, 0])
cov2 = np.diag(np.ones(2))
X1 = np.random.multivariate_normal(mus2, cov2, size=n)
Y1 = np.array([1]*n)
```

```python
mus2 = np.array([6, 0])
cov2 = np.diag(np.ones(2)*0.1)
X2 = np.random.multivariate_normal(mus2, cov2, size=n)
Y2 = np.array([2]*n)
```

```python
mus3 = np.array([8, 8])
cov3 = np.diag(np.ones(2)*0.1)
X3 = np.random.multivariate_normal(mus3, cov3, size=n)
Y3 = np.array([3]*n)
```

```python
X = np.concatenate([X1, X2, X3])
Y = np.concatenate([Y1, Y2, Y3])

print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0], X[:, 1], c=Y)
```

```python
neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(X, Y)
```

```python
score = neigh.score(X, Y)
```
